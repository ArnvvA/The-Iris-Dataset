# Modeling- What is our baseline:
what is the simplest model we can think of?
->In this case: baseline= randomly guessing a species. So expected accuracy therefore = 33% (3 possible classes with equal distribution(50 each))
### Therefore, baseline =33%

# A simple manual model
Lets manually look at data and make some cutoffs

```python
def single_feature_prediction(petal_length):
    ''' predicts iris flower species given a petal length'''
    if petal_length <2.5:
        return 0
    elif petal_length <4.75:
        return 1
    else:
        return 2
```

#### The other two are not as simple as there is no clear cutoff(?)
->we can take some liberties where the cutoffs are not as clear and accept that there will be some rare outliers that end up on the wrong side of the cutoff

```python
x_train[:,2]
```

```python
manual_y_predictions = np.array([single_feature_prediction(val) for val in x_train[:,2]])
```

```python
manual_y_predictions
```

#### Now to check how accurate that is
using np.mean to get a percentage value

```python
manual_model_accuracy = np.mean(manual_y_predictions == y_train)
```

```python
print(f"Manual model accuracy: {manual_model_accuracy * 100:.2f}%")
```

## Using cross-validation for evaluation

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model, x_train, y_train, cv = 5, scoring = 'accuracy') 
#cv = nmuber of splits/iterations
```

# Where are we misclassifying 
what is causing the inacuracies

```python
y_pred = cross_val_predict(model, x_train, y_train, cv = 5)
```

```python
predicted_correctly_mask = y_pred == y_train
```

```python
y_pred == y_train
```

```python
not_predicted_correctly =  ~predicted_correctly_mask
```

```python
x_train[not_predicted_correctly]
```

```python
df_predictions = df_train.copy()
```

```python
df_predictions['correct_prediction'] = predicted_correctly_mask
```

```python
df_predictions['prediction'] = y_pred
```

```python
df_predictions['prediction_label'] = df_predictions['prediction'].map({0:'setosa',1:'versicolor',2:'virginica'})
```

```python
df_predictions[not_predicted_correctly]
```

```python
sns.scatterplot(x = 'petal length (cm)', y = 'petal width (cm)', data = df_predictions , hue='prediction_label')
```

```python
sns.scatterplot(x = 'petal length (cm)', y = 'petal width (cm)', data = df_predictions , hue='target_name')
```

```python
def plot_incorrect_predictions(df_predictions,x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2,2, figsize = (10, 10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, data= df_predictions, hue = 'prediction_label', ax= axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, data= df_predictions, hue = 'target_name', ax= axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, data= df_predictions, hue = 'correct_prediction', ax= axs[2])
    axs[3].set_visible(False)
```

```python
plot_incorrect_predictions(df_predictions,'petal length (cm)','petal width (cm)')
```

# Model Tuning
Model Tuning is trying to determine the parameters of your model that maximise model performance (also known as 'hyperparameters')

```python
for reg_param in [1, 1.3, 1.9, 2,10, 15,100]:
    print(reg_param)
    
    model = LogisticRegression(max_iter=200, C=reg_param)
    accuracies = cross_val_score(model, x_train, y_train, cv = 5, scoring = 'accuracy')
    print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
```

# Final Model

```python
model = LogisticRegression(max_iter=200, C=15)
```

# Checking our model on the Test Set

```python
x_test=df_test.drop(columns=['target','target_name']).values
y_test= df_test['target'].values
```

```python
y_test
```

## Training our final model on our full training data

```python
model.fit(x_train, y_train)
```

```python
y_test_pred = model.predict(x_test)
```

```python
test_set_correct_prediction = y_test_pred == y_test
test_set_accuracy = np.mean(test_set_correct_prediction)
print(f'Test set accuracy: {test_set_accuracy *100:2f}%')
```

```python
df_predictions_test = df_test.copy()
df_predictions_test['correct_prediction'] = test_set_correct_prediction
df_predictions_test['prediction'] = y_test_pred
df_predictions_test['prediction_label'] = df_predictions_test['prediction'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
```

```python
plot_incorrect_predictions(df_predictions_test, 'petal length (cm)', 'petal width (cm)')
```
